```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import tensordot
from tensorflow import keras
from tensorflow.keras import layers

# Load data
books_data = pd.read_csv("data/Books.csv")
ratings_data = pd.read_csv("data/Ratings.csv")
users_data = pd.read_csv("data/Users.csv")

# Drop unnecessary columns
books_data.drop(labels=['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1, inplace=True)

# Remove rows with incorrect 'Year-Of-Publication'
invalid_years = (books_data['Year-Of-Publication'] == 'DK Publishing Inc') | (books_data['Year-Of-Publication'] == 'Gallimard')
books_data = books_data.drop(books_data[invalid_years].index)

# Convert 'Year-Of-Publication' to integer
books_data['Year-Of-Publication'] = books_data['Year-Of-Publication'].astype(int)

# Display top 10 authors by number of books
author_counts = books_data.groupby('Book-Author')['Book-Title'].count()
sorted_authors = author_counts.sort_values(ascending=False)
top_authors = sorted_authors.head(10)

plt.figure(figsize=(12, 6))
top_authors.plot(kind='bar')
plt.xlabel('Author Name')
plt.ylabel('Number of Books')
plt.title('Top 10 Authors by Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display sorted ratings
sorted_ratings = ratings_data['Book-Rating'].value_counts().sort_index()
pd.DataFrame({'Book-Rating': sorted_ratings.index, 'Sum': sorted_ratings.values})

# Merge ratings and books data
merged_data = pd.merge(ratings_data, books_data, on='ISBN', how='left')
grouped_books = merged_data.groupby('ISBN').sum()

# User-ID encoding
user_ids = ratings_data['User-ID'].unique().tolist()
user_to_encoded = {x: i for i, x in enumerate(user_ids)}
encoded_to_user = {i: x for i, x in enumerate(user_ids)}

# ISBN encoding
isbn_ids = ratings_data['ISBN'].unique().tolist()
isbn_to_encoded = {x: i for i, x in enumerate(isbn_ids)}
encoded_to_isbn = {i: x for i, x in enumerate(isbn_ids)}

# Map User-ID and ISBN to ratings dataframe
ratings_data['user'] = ratings_data['User-ID'].map(user_to_encoded)
ratings_data['book_title'] = ratings_data['ISBN'].map(isbn_to_encoded)

# Get the number of users and book titles
num_users = len(user_to_encoded)
num_book_titles = len(isbn_to_encoded)

# Convert rating to float
ratings_data['Book-Rating'] = ratings_data['Book-Rating'].values.astype(np.float32)

# Min and max rating values
min_rating = min(ratings_data['Book-Rating'])
max_rating = max(ratings_data['Book-Rating'])

print('Number of Users: {}, Number of Book Titles: {}, Min Rating: {}, Max Rating: {}'.format(
    num_users, num_book_titles, min_rating, max_rating
))

# Clean books data
clean_books = books_data.dropna()

# Sort books by ISBN
sorted_books = clean_books.sort_values('ISBN', ascending=True)
unique_books = sorted_books.drop_duplicates('ISBN')

# Prepare data for recommender
isbn_list = unique_books['ISBN'].tolist()
book_title_list = unique_books['Book-Title'].tolist()
book_author_list = unique_books['Book-Author'].tolist()
year_of_publication_list = unique_books['Year-Of-Publication'].tolist()
publisher_list = unique_books['Publisher'].tolist()

# Create a new DataFrame
books_new = pd.DataFrame({
    'isbn': isbn_list,
    'book_title': book_title_list,
    'book_author': book_author_list,
    'year_of_publication': year_of_publication_list,
    'publisher': publisher_list
})

# Take a subset of the data for training
books_new = books_new[:20000]
ratings_data = ratings_data.sample(frac=1, random_state=42)

# Create input features (x) and target variable (y)
x = ratings_data[['user', 'book_title']].values
y = ratings_data['Book-Rating'].apply(lambda x: (x - min_rating) / (max_rating - min_rating)).values

# Split the data into train and validation sets
train_indices = int(0.9 * ratings_data.shape[0])
x_train, x_val, y_train, y_val = (
    x[:train_indices],
    x[train_indices:],
    y[:train_indices],
    y[train_indices:]
)

# Define the recommender model
class CustomRecommender(keras.Model):

    def __init__(self, num_users, num_book_titles, embedding_size, dropout_rate=0.2, **kwargs):
        super(CustomRecommender, self).__init__(**kwargs)
        self.num_users = num_users
        self.num_book_titles = num_book_titles
        self.embedding_size = embedding_size
        self.dropout_rate = dropout_rate

        self.user_embedding = layers.Embedding(
            num_users,
            embedding_size,
            embeddings_initializer='he_normal',
            embeddings_regularizer=keras.regularizers.l2(1e-6)
        )
        self.user_bias = layers.Embedding(num_users, 1)

        self.book_title_embedding = layers.Embedding(
            num_book_titles,
            embedding_size,
            embeddings_initializer='he_normal',
            embeddings_regularizer=keras.regularizers.l2(1e-6)
        )
        self.book_title_bias = layers.Embedding(num_book_titles, 1)

        self.dropout = layers.Dropout(rate=dropout_rate)

    def call(self, inputs):
        user_vector = self.user_embedding(inputs[:, 0])
        user_vector = self.dropout(user_vector)
        user_bias = self.user_bias(inputs[:, 0])

        book_title_vector = self.book_title_embedding(inputs[:, 1])
        book_title_vector = self.dropout(book_title_vector)
        book_title_bias = self.book_title_bias(inputs[:, 1])

        dot_user_book_title = tensordot(user_vector, book_title_vector, 2)
        x = dot_user_book_title + user_bias + book_title_bias

        return tf.nn.sigmoid(x)

# Initialize and compile the model
recommender_model = CustomRecommender(num_users, num_book_titles, 50)
recommender_model.compile(
    loss=keras.losses.BinaryCrossentropy(),
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    metrics=[keras.metrics.RootMeanSquaredError()]
)


```

```python
# Train the model
history = recommender_model.fit(
    x=x_train,
    y=y_train,
    batch_size=64,
    epochs=50,
    validation_data=(x_val, y_val)
)

# Generate book recommendations for a random user
user_id = ratings_data['User-ID'].sample(1).iloc[0]
books_read_by_user = ratings_data[ratings_data['User-ID'] == user_id]

# Generate book recommendations for a random user (continued)
books_not_read_by_user = books_new[~books_new['isbn'].isin(books_read_by_user['ISBN'].values)]['isbn']
books_not_read_by_user = list(
    set(books_not_read_by_user)
    .intersection(set(isbn_to_encoded.keys()))
)

books_not_read_by_user = [[isbn_to_encoded.get(x)] for x in books_not_read_by_user]
user_encoder = user_to_encoded.get(user_id)
user_book_array = np.hstack(
    ([[user_encoder]] * len(books_not_read_by_user), books_not_read_by_user)
)
ratings_predictions = recommender_model.predict(user_book_array).flatten()

top_ratings_indices = ratings_predictions.argsort()[-10:][::-1]

recommended_book_ids = [
    encoded_to_isbn.get(books_not_read_by_user[x][0]) for x in top_ratings_indices
]

top_books_read_by_user = (
    books_read_by_user.sort_values(
        by='Book-Rating',
        ascending=False
    )
    .head(10)['ISBN'].values
)

recommended_books_data = books_new[books_new['isbn'].isin(recommended_book_ids)]

# Display book recommendations in DataFrame form
recommended_books_dataframe = []
for row in recommended_books_data.itertuples():
    recommended_books_dataframe.append([row.book_title, row.book_author])

books_read_by_user_dataframe = []
for row in books_read_by_user.itertuples():
    books_read_by_user_dataframe.append([row.book_title, row.book_author])

# Create DataFrames for output
output_columns = ['Book Title', 'Book Author']
df_books_read_by_user = pd.DataFrame(books_read_by_user_dataframe, columns=output_columns)
df_recommended_books = pd.DataFrame(recommended_books_dataframe, columns=output_columns)

# Display recommendation results in DataFrame form
print("Showing recommendations for user: {}".format(user_id))
print("===" * 9)
print("Books with high ratings from user")
print("----" * 8)
print(df_books_read_by_user)
print("----" * 8)
print("Top 10 recommended books")
print("----" * 8)
df_recommended_books
```
